### 1. Compute statistics on a Python list
Given the following list

````python
l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]
# Write your solution here

mean = sum(l)/len(l)
print('Mean is {}'.format(mean))

median = sorted(l)[len(l)//2]
print('Median is {}'.format(median))

var = sum([(el-mean)**2 for el in l])/len(l)
print('Var is {}'.format(var))
std = var**(1/2)
print('Std is {}'.format(std))
````

### 5. How to create a new column from existing columns of a numpy array

Create a new column for the flower volume V in `iris`, where volume is computing according to this formula (the assumption here is that the shape of the iris flower is approximately conical):

$$ Volume = \frac{\pi \times Sepal Length^2 \times Petal Length}{3}$$

````python
# Input
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')

# Write your solution here

# Compute volume
# ≠print(iris_2d[0, 0], float(iris_2d[0, 0]), np.pi * 5.1**2 * 1.4 / 3  )

sepallength = iris_2d[:, 0].astype('float')
petallength = iris_2d[:, 2].astype('float')
volume = np.pi * petallength * (sepallength**2)/3

# Introduce new dimension to match iris_2d's
volume = volume[:, np.newaxis]

# Add the new column
out = np.hstack([iris_2d, volume])

# View
out[:4]
#> array([[b'5.1', b'3.5', b'1.4', b'0.2', b'Iris-setosa', 38.13265162927291],
#>        [b'4.9', b'3.0', b'1.4', b'0.2', b'Iris-setosa', 35.200498485922445],
#>        [b'4.7', b'3.2', b'1.3', b'0.2', b'Iris-setosa', 30.0723720777127],
#>        [b'4.6', b'3.1', b'1.5', b'0.2', b'Iris-setosa', 33.238050274980004]], dtype=object)
````

### 12. Compute an estimate the pressure exercised by each car. 

Append the result as a column named 'Pressure [Pa]' after the column weight of `df`. Pressure *P*, expressed in Pascal (Pa) can be computed with this formula.

$$P =  \frac{F}{A}$$

Where the Surface Area *A*, expressed in square metres, can be computed with this formula (we hypothesise that the car is roughly rectangular in shape, measure units in square brackets):

$$A = Width [m] \times Length [m]$$ 

And the applied force *F*, expressed in Newton, can be computed with this formula, measure units in square brackets:

$$F = Weight [kg] \times g$$,

where g is the gravitational acceleration, that is: 

$$ g = 9.81 \frac{m}{s^2}$$


````python
# Write your solution herea
g = 9.81
area = (df_renamed["Width"] / 100) * (df_renamed["Length"] / 100)
applied_force = g * df_renamed["Weight"]
pressure = applied_force / area
df_renamed["Pressure (Pa)"] = pressure
df_renamed.head()
pressure
````

The measure unit of pressure is Pascal which can be a bit unwieldy. Convert the measures of pressure to millibars 
where 1 millibar = 10^2 Pa

````python
df_renamed["Pressure (millibar)"] = pressure / 10**2
````
