### 1. Compute statistics on a Python list
Given the following list
```
l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]
```
Compute the mean, the median and the standard deviation of the values in the list, using only built in functionalities of python (no imported libraries, no NumPy or Pandas)

````python
l = [12, 45, 12, 999, 10, 8, 76, 20, 10, 10, 7, 70, 17]

# Write your solution here

# mean
mean = sum(l)/len(l)
print("mean =", mean)

#median
l.sort()
print("sorted =", l)

if len(l)%2 == 0:
    median = (l[(len(l)//2)-1] + l[(len(l)//2)])/2
else:
    median = l[(len(l)//2)-1]
    
print("median =",median)

squared_difference = [ (x - mean)**2 for x in l ]

pvar = (sum(squared_difference)/(len(l)))
print("population variance =",pvar)
pstd = (sum(squared_difference)/(len(l)))**(1/2)
print("population standard deviation =", pstd)
````

For these exercises we will work on the IRIS dataset, a very popular multivariate dataset for data anaylisis, first introduced by Ronald Fisher in 1936.

See more about the IRIS dataset heare: https://archive.ics.uci.edu/ml/datasets/Iris

The data source is the file `iris.data` retrieved from the Web. It contains the data for this example in comma separated values (CSV) format. The number of columns is 5 and the number of rows is 150. The data set is composed of 50 samples from each of three species of Iris Flower: Iris Setosa, Iris Virginica and Iris Versicolor. Four features were measured from each sample: the length and the width of the sepals and petals, in centimetres. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species one from each other.

The variables are:

    sepal_length: Sepal length, in centimeters.
    sepal_width: Sepal width, in centimeters.
    petal_length: Petal length, in centimeters.

### 12. Compute an estimate the pressure exercised by each car. 

Append the result as a column named 'Pressure [Pa]' after the column weight of `df`. Pressure *P*, expressed in Pascal (Pa) can be computed with this formula.

$$P =  \frac{F}{A}$$

Where the Surface Area *A*, expressed in square metres, can be computed with this formula (we hypothesise that the car is roughly rectangular in shape, measure units in square brackets):

$$A = Width [m] \times Length [m]$$ 

And the applied force *F*, expressed in Newton, can be computed with this formula, measure units in square brackets:

$$F = Weight [kg] \times g$$,

where g is the gravitational acceleration, that is: 

$$ g = 9.81 \frac{m}{s^2}$$


````python
# Write your solution herea
width_meters  =  (df['Width']*2.54)/100
length_meters = (df['Length']*2.54)/100
weight_kilograms = df['Weight']/2.205

surface_area_meters = width_meters*length_meters
force_newtons = weight_kilograms*9.81
pressure_pascal = force_newtons/surface_area_meters
df.insert(df.columns.get_loc('Origin'), 'Pressure [Pa]', pressure_pascal)
display(df)
````

The measure unit of pressure is Pascal which can be a bit unwieldy. Convert the measures of pressure to millibars 
where 1 millibar = 10^2 Pa

````python
## Write your solution here
millibar = df['Pressure [Pa]']/100
df = df.rename(columns={'Pressure [Pa]': 'Pressure [mbar]'})
df['Pressure [mbar]'] = millibar
display(df)
````
